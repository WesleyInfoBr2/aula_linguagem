Exercício 3) Crie 4 variáveis e faça receber os valores das quatro colunas (Passivo Circulante, Passivo Não Circulante, Ativo Circulante e Ativo Não Circulante). Depois calcule o grau de endividamento geral e apresente o resultado na tela com a seguinte mensagem: "Grau de endividamento geral: "

Dica: o endividamento geral pode ser calculado da seguinte forma = ( (Passivo Circulante + Passivo Não Circulante) / (Ativo Circulante + Ativo Não Circulante) ) * 100

(peso = 1,5)

```python
AC = df["Ativo Circulante"].sum()
ANC = df["Ativo não circulante"].sum()
PC = df["Passivo Circulante"].sum()
PNC = df["Passivo não circulante"].sum()

EndivGeral = ( ( PC + PNC) / ( AC + ANC) ) * 100
print("Grau de endividamento geral:", round(EndivGeral, 2), "%")
```

Exercício 4) Aparentemente o cálculo do Patrimônio líquido foi realizado de forma errada e é preciso corrigir. Utilize as funções map e lambda para aplicar desconto de 8% nos valores do Patrimônio líquido e guarde os novos valores em uma lista.  (peso = 1,5)

Exercício 5) Guarde os valores do Lucro Líquido da BRF em uma lista. Utilize as funções reduce e lambda para apresentar o Lucro Líquido acumulado.  (peso = 1,5)
